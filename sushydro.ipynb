# Direct Employment Generation

```python
def direct_employment(num_direct_employees, num_months, energy_per_month):
    return (num_direct_employees * num_months) / energy_per_month

dir_emp_example = direct_employment(150, 5, 5000)
print(dir_emp_example)
```

# Hydrological Resource Monitoring

```python
def hydro_resources(flow_measurements, time):
    return flow_measurements / time

hyd_res_example = hydro_resources(900, 30)
print(hyd_res_example)
```

# Normalized Indicator Values

## For member functions increasing with indicator q(x_i)

```python
def q_j(x_j, min_j, max_j):
    result = 0
    
    if x_j <= min_j:
        pass
    elif min_j < x_j and x_j <= max_j:
        result = x_j - min_j
    elif x_j == max_j:
        result = max_j - min_j
    else:
        result = 1
    return result

example_qj = q_j(20, 11, 22)
print(example_qj)
```

## For member functions decreasing with indicator q(x_i)

```python
def q_i(x_i, min_i, max_i):
    result = 0
    if x_i <= min_i:
        pass
    elif min_i < x_i and x_i <= max_i:
        result = max_i -x_i
    elif x_i == max_i:
        result = max_i - min_i
    else:
        result = 1
    return result

example_qi = q_i(20, 11, 12)
print(example_qi)
```

# Single Preference Index

```python
def q_n(w_n, q_n): # assuming w_n and q_n are not matrices
    return w_n * q_n

example_qn = q_n(50, 30)
print(example_qn)
```

# Sustainability Index of Specific Criteria

```python
def q_n_specific(w_vect, q_vect):
    if not len(w_vect) == len(q_vect):
        raise("Vector size mismatch for w and q in input")
    else:
        sum = 0
        for i in range(len(w_vect)):
            sum += w_vect[i] * q_vect[i]
        return sum

example_q_n_specific = q_n_specific([11, 12, 13], [9, 10, 11])
print(example_q_n_specific)
```
